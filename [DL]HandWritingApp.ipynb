```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

fileList=glob('c:/data/mnist_img/*.jpg')
print(fileList)

for x in fileList:
    # conver('L')은 흑백으로
    img = Image.open(x).resize((28, 28)).convert('L') 
    ## TODO 배경 변경 [255 - img는 흰배경 -> 검정]
    img = 255-np.array(img)
    # scaling
    img=img/255
    np.shape(img)

    plt.figure(figsize=(3,3))
    plt.imshow(img, cmap= 'gray')
    plt.show()
    # 모델제작할 때 가장 아래쪽 레이어 10이라는 숫자에 의해
    # 10개의 확률값이 생성, 가장 큰 확률 값을 갖고 있는 위치 값
    P_num = model.predict(img.reshape(-1,28,28))
    # 그 위치 값을 숫자로
    np.argmax(P_num)
```
